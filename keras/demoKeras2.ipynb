el problema XOR es famoso porque una red neuronal de una sola capa (un perceptrón simple) no puede resolverlo, ya que no es un problema "linealmente separable". Para resolverlo, se necesita al menos una "capa oculta" de neuronas, que es exactamente lo que hace tu código.

Parámetros ($Weights$ y $Biases$): Cada conexión tiene un peso y cada neurona un sesgo. La fórmula es $Par\acute{a}metros = (Entradas \times Neuronas) + Sesgos$. Aquí: $(2 \times 16) + 16 = 48$.

```python
# Ejemplo 2: Comportamiento puertas XOR
#  importamos las clases que utilizaremos
import numpy as np
from keras.models import Sequential
from keras.layers import Dense

# creamos los arrays de entrada y salida.
# cargamos las 4 combinaciones de las compuertas XOR
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")

# y estos son los resultados que se obtienen, en el mismo orden
target_data = np.array([[0],[1],[1],[0]], "float32")

# Creación del modelo de red neuronal
model = Sequential()
# la dimension viene dada por datos de entrada

model.add(Dense(16, input_dim=2, activation='relu')) # Parámetros: 16 (num neuronas) x 2 (dimensiones) + 16(sesgo) = 48
model.add(Dense(1, activation='sigmoid')) # 1 (neuronas)*16(neuronas capa anterior) + 1(sesgo)=17

# printamos el resumen del modelo de red neuronal
print(model.summary())

# Con compile podemos ajustar nuestro modelo
# como le paso en el compile esos dos el loss y el binary me los guarda en el score 
model.compile(loss='mean_squared_error',
            optimizer='adam',
            metrics=['binary_accuracy'])

# entrenaremos la red recibiendo por parámetro los arrays de entrada y salida, así como el número de iteraciones de aprendizaje (epochs) de entrenamiento
print("Comenzando entrenamiento...")
model.fit(training_data, target_data, epochs=1000, verbose=False)
print("Modelo entrenado!")

# evaluamos el modelo
scores = model.evaluate(training_data, target_data)

#probamos a hacer una predicción
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
print (model.predict(training_data).round())
```
